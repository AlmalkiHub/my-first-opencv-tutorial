#### **3) Display Images** 

```python
# Load an images using "imread" function after specifying the image path
image = cv2.imread("/content/images/church_iceland.jpeg")
```

```python
print(image)
print(type(image))
print(image.shape)
```

```python
# Display the loaded image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

#### Helper function to display images 

```python
def display_image(title="", image=None):
  """
  This function takes an image as an input and returns a displayed image.
  """

  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
# Load my beautiful image 
ali_image = cv2.imread("/content/083A5105.jpg")
```

```python
# Let's use now the helper function to display an image 
display_image("Display my beautiful image", ali_image)
```
